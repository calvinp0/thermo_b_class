```python
import pandas as pd
import matplotlib.pyplot as plt
# use ggplot style for more sophisticated visuals
plt.style.use('ggplot')
```

```python
# Import files and create columns x and y, the first row is data
data_bp = pd.read_csv('/Users/calvin/Downloads/Default Dataset.csv', names=['x', 'y'])
data_bp
```

```python
data_gr = pd.read_csv('/Users/calvin/Downloads/Dataset 1.csv' , names=['x', 'y'])
data_gr
```

```python
data_red = pd.read_csv('/Users/calvin/Downloads/Dataset 2.csv', names=['x', 'y'])
```

```python
# Plot the data as lines
# Y is pressure
# X is volume
# Add X mark at the end point of BP
plt.scatter(data_bp['x'].iat[-1], data_bp['y'].iat[-1], marker='x', color='black', zorder=10)
plt.plot(data_bp['x'], data_bp['y'], 'b-', label='BP')
plt.plot(data_gr['x'], data_gr['y'], 'r-', label='GR')
plt.plot(data_red['x'], data_red['y'], 'g-', label='RED')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.title('Pressure-Volume Diagram: Regions of Phase Equilibrium')

# Place text box in middle start of the plot
plt.text(0.06, 0.7, 'Liquid \n Region',horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, wrap=True)
# Supercritical region
plt.text(0.5, 0.7, 'Supercritical \n Region',horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, wrap=True)
# Liquid Vapour Two Phase Region
plt.text(0.2, 0.3, 'Liquid-Vapour \n Two Phase Region',horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, wrap=True)
# vapour region
plt.text(0.6, 0.25, 'Vapour \n Region',horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, wrap=True)
# gas Region
plt.text(0.9, 0.4, 'Gas \n Region',horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, wrap=True)
# Create a dotted line at the X mark going horizontally across the X axis - the critical point,
plt.axhline(y=data_bp['y'].iat[-1], color='black', linestyle='dashed', alpha=0.5)

# Arrow pointing to the X mark with a callout being 'Critical Point', the arrow has an alpha of 0.5 and is thin and comes from the top right
plt.annotate('Critical Point', xy=(data_bp['x'].iat[-1], data_bp['y'].iat[-1]), xytext=(150, 400), arrowprops=dict(facecolor='black', shrink=0.05, alpha=0.5, width=0.5, headwidth=5), horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
             
#plt.show()
plt.savefig('PvVDia.png', dpi=300)
```
