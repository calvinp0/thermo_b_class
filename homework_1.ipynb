## Question 2

### Compute the variable $X = \sum_{i=1}^{N} x_i$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the three distributions
distributions = [
    {'name': 'Uniform', 'dist': np.random.uniform},
    {'name': 'Normal', 'dist': np.random.normal},
    {'name': 'Exponential', 'dist': np.random.exponential}
]

# Define the sample sizes to test
Ns = [10, 30, 100]

# Loop over the distributions
for dist in distributions:
    # Loop over the sample sizes
    for N in Ns:
        # Generate N random numbers from the distribution
        x = dist['dist'](size=N)
        # Compute X
        X = np.sum(x)
        # Print the result
        print(f"{dist['name']} distribution, N={N}: X={X}")

        # Plot a histogram of the x values
        plt.hist(x, bins=20, density=True, alpha=0.5, label=f"N={N}")

    # Add a legend to the plot
    plt.legend()
    plt.title(f"{dist['name']} distribution")
    plt.show()
```

## Question 3

### Repeat 1500 times, and then plot a histogram of X from the generated data

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the three distributions
distributions = [
    {'name': 'Uniform', 'dist': np.random.uniform},
    {'name': 'Normal', 'dist': np.random.normal},
    {'name': 'Exponential', 'dist': np.random.exponential}
]

# Define the sample sizes to test
Ns = [10, 30, 100]

# Number of simulations to run
num_sims = 1500

# Initialize array to store X values
X_vals = np.zeros((len(distributions), len(Ns), num_sims))

# Loop over the distributions
for i, dist in enumerate(distributions):
    # Loop over the sample sizes
    for j, N in enumerate(Ns):
        # Loop over the number of simulations
        for k in range(num_sims):
            # Generate N random numbers from the distribution
            x = dist['dist'](size=N)
            # Compute X
            X = np.sum(x)
            # Store the X value in the array
            X_vals[i, j, k] = X

# Plot histograms of X values
for i, dist in enumerate(distributions):
    for j, N in enumerate(Ns):
        plt.hist(X_vals[i, j, :], bins=20, density=True, alpha=0.5, label=f"N={N}")
    plt.legend()
    plt.title(f"{dist['name']} distribution")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, expon, norm

# Define the distributions
distributions = [
    {'name': 'Uniform', 'dist': uniform(loc=0.0, scale=1.0)},
    {'name': 'Exponential', 'dist': expon(scale=1.0)},
    {'name': 'Normal', 'dist': norm(loc=0.0, scale=1.0)}
]

# Define the sample sizes
Ns = [10, 30, 100]

# Define the number of iterations
num_iter = 1500

# Initialize an array to store the values of X
X_vals = np.zeros((len(distributions), len(Ns), num_iter))

# Loop over the distributions
for i, dist in enumerate(distributions):
    # Compute the mean and standard deviation of the distribution
    mu = dist['dist'].mean()
    sigma = dist['dist'].std()
    # Loop over the sample sizes
    for j, N in enumerate(Ns):
        # Generate N random numbers from the distribution for each iteration
        x = dist['dist'].rvs(size=(num_iter, N))
        # Compute X for each iteration
        X = np.sum(x, axis=1)
        # Store the values of X
        X_vals[i, j, :] = X

# Plot the histograms of X for each combination of distribution and sample size
for i, dist in enumerate(distributions):
    plt.figure()
    plt.title(f"{dist['name']} Distribution")
    for j, N in enumerate(Ns):
        plt.hist(X_vals[i, j, :], bins=20, density=True, alpha=0.5, label=f"N={N}")
    # Predicted distribution based on Central Limit Theorem
    mu = Ns[-1] * mu
    sigma = np.sqrt(Ns[-1]) * sigma
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
    plt.plot(x, norm.pdf(x, loc=mu, scale=sigma), label="CLT Prediction")
    plt.legend()
    plt.show()
```
